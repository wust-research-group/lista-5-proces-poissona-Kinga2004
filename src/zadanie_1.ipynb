```python
import numpy as np

def poisson(lambd, T):
    t = 0
    N = 0
    trajectory = []

    while t < T:
        
        waiting_time = np.random.exponential(scale=1/lambd)
        t += waiting_time

        if t < T:
            
            N += 1
            trajectory.append(t)

    return np.array(trajectory)
```

```python
import matplotlib.pyplot as plt

lambd = 0.5  
T = 10       

trajectory = poisson(lambd, T)

plt.figure(figsize=(10, 5))
plt.step(trajectory, range(len(trajectory)), where='post')
plt.xlabel('Czas')
plt.ylabel('Liczba zdarze≈Ñ')
plt.title('Trajektoria procesu Poissona')
plt.show()
```

```python
def poisson_distribution(k, lambd, T):
    return (lambd * T) ** k / (np.math.factorial(k) * np.exp(lambd * T))

k_values = np.arange(0, 10) 
lambd = 0.5
T = 10

poisson_values = [poisson_distribution(k, lambd, T) for k in k_values]

plt.bar(k_values, poisson_values)
plt.xlabel('Liczba zdarze≈Ñ')
plt.ylabel('Prawdopodobie≈Ñstwo')
plt.title('Rozk≈Çad Poissona dla procesu Poissona')
plt.show()
```
